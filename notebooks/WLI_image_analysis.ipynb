```python
csv_paths = list_csvs("../data/WLI/")
print_list(csv_paths)

output_folder = "../data/WLI/output/"
makedirs(output_folder)
```

```python
paramss = []
bin_step = 1

for csv_path in csv_paths:
    df = pd.read_csv(csv_path, skiprows=19, encoding="Shift-JIS", header=None) * 10000 # um -> angstrom
    head = pd.read_csv(csv_path, nrows=19, encoding='Shift-JIS', index_col=0, usecols=[0,1]).T
    pixel_size = float(head["XYキャリブレーション"][0]) / 1000 # nm -> um
    display(head)

    imshow(df.values, cmap="gray", title=f"{csv_path}")
    plt.show()
    imwrite_int16(join(output_folder, basename(csv_path.replace(".csv", ".tif"))), df.values, pixel_size_um=pixel_size)
    d_design = float(csv_path.split("_d")[1].split("nm")[0])

    df_flat = df.values.flatten()
    df_unique = np.unique(df_flat)
    df_max = df_unique.max()
    df_min = df_unique.min()
    df_step = round(np.abs(np.diff(df_unique)).min())*bin_step
    bins = np.arange(df_min-0.5, df_max+0.5, df_step)

    # fig, ax = plt.subplots(figsize=(5, 5))
    fig, ax = plt.subplots()
    params = plot_histogram_with_multi_gaussian_fit(df_flat, 2, initial_guess=[0.5, -d_design*10, 20, 0.5, 0, 20], bins=bins, ax=ax)
    params = np.array(params).reshape(-1, 3)
    print(params)
    paramss.append(params)

    plt.xlabel(r"Height [$\mathrm{\AA}$]")
    plt.ylabel(r"Probability density [/$\mathrm{\AA}$]")

    # plt.ylim(1/df.size, df_max)
    # plt.yscale("log")

    plt.tight_layout()
    plt.savefig(join(output_folder, basename(csv_path).replace(".csv", "_histogram.svg")))
    plt.show()
```
